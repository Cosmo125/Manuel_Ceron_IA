## 1. Los Orígenes (1940s-1950s)

### Antecedentes Filosóficos
La idea de crear máquinas inteligentes tiene raíces profundas en la filosofía y la mitología:
- **Aristóteles** (384-322 a.C.) desarrolló el silogismo, precursor de la lógica formal
- **Ramon Llull** (1232-1316) creó la "Ars Magna", un sistema mecánico para generar ideas
- **Gottfried Leibniz** (1646-1716) soñó con un lenguaje universal del pensamiento

### Los Fundadores Modernos
- **Alan Turing** (1912-1954): Propuso la "Máquina de Turing" y el famoso "Test de Turing"
- **John von Neumann** (1903-1957): Desarrolló la arquitectura de computadoras que lleva su nombre
- **Norbert Wiener** (1894-1964): Fundó la cibernética, estudiando el control y comunicación

## 2. La Era Dorada (1956-1974)

### La Conferencia de Dartmouth (1956)
Considerada el momento fundacional de la IA como disciplina académica:
- **Organizadores**: John McCarthy, Marvin Minsky, Nathaniel Rochester, Claude Shannon
- **Objetivo**: "Proceder sobre la base de la conjetura de que cada aspecto del aprendizaje o cualquier otra característica de la inteligencia puede, en principio, ser descrita con tanta precisión que una máquina puede ser hecha para simularlo"

### Primeros Programas de IA
- **Logic Theorist** (1955): Probaba teoremas matemáticos
- **General Problem Solver** (1957): Resolvía problemas mediante búsqueda heurística
- **ELIZA** (1965): Primer programa de procesamiento de lenguaje natural

## 3. El Primer Invierno de la IA (1974-1980)

### Causas del Estancamiento
- **Limitaciones computacionales**: Hardware insuficiente para las ambiciones
- **Problemas de escalabilidad**: Los enfoques no funcionaban en problemas reales
- **Reducción de financiamiento**: Gobierno y empresas perdieron confianza

### Lecciones Aprendidas
- La importancia del conocimiento específico del dominio
- Necesidad de enfoques más pragmáticos
- Reconocimiento de la complejidad computacional

## 4. Sistemas Expertos y Renacimiento (1980-1987)

### El Auge de los Sistemas Expertos
- **DENDRAL**: Análisis de estructuras moleculares
- **MYCIN**: Diagnóstico médico de infecciones
- **XCON**: Configuración de sistemas informáticos

### Características de los Sistemas Expertos
- Base de conocimiento separada del motor de inferencia
- Reglas if-then para representar conocimiento
- Capacidad de explicar su razonamiento

```python
# Ejemplo simple de un sistema experto básico
class SistemaExpertoSimple:
    def __init__(self):
        self.base_conocimiento = {
            'fiebre_alta': ['infeccion', 'virus'],
            'dolor_garganta': ['resfriado', 'faringitis'],
            'tos_seca': ['alergia', 'resfriado'],
            'congestion_nasal': ['resfriado', 'alergia']
        }
    
    def diagnosticar(self, sintomas):
        posibles_diagnosticos = set()
        
        for sintoma in sintomas:
            if sintoma in self.base_conocimiento:
                posibles_diagnosticos.update(self.base_conocimiento[sintoma])
        
        return list(posibles_diagnosticos)

# Ejemplo de uso
sistema = SistemaExpertoSimple()
sintomas_paciente = ['fiebre_alta', 'dolor_garganta']
diagnostico = sistema.diagnosticar(sintomas_paciente)

print(f"Síntomas: {sintomas_paciente}")
print(f"Posibles diagnósticos: {diagnostico}")
```

## 5. Segundo Invierno de la IA (1987-1993)

### Nuevos Desafíos
- **Limitaciones de los sistemas expertos**: Frágiles y difíciles de mantener
- **Competencia de las PCs**: Hardware más barato y accesible
- **Expectativas no cumplidas**: Promesas exageradas no realizadas

### Desarrollos Paralelos
- Investigación continuó en universidades
- Desarrollo de redes neuronales (backpropagation)
- Avances en robótica y visión por computadora

## 6. El Surgimiento del Machine Learning (1993-2011)

### Cambio de Paradigma
- De sistemas basados en reglas a sistemas que aprenden de datos
- Enfoque estadístico y probabilístico
- Mayor disponibilidad de datos y poder computacional

### Hitos Importantes
- **1997**: Deep Blue vence a Garry Kasparov en ajedrez
- **2005**: Vehículos autónomos en el DARPA Grand Challenge
- **2011**: Watson de IBM gana en Jeopardy!

## 7. La Era del Deep Learning (2012-presente)

### Factores del Renacimiento
- **Big Data**: Disponibilidad masiva de datos
- **GPU Computing**: Procesamiento paralelo masivo
- **Avances algorítmicos**: Mejores técnicas de entrenamiento

### Logros Revolucionarios
- **2012**: AlexNet revoluciona la visión por computadora
- **2016**: AlphaGo vence al campeón mundial de Go
- **2020**: GPT-3 demuestra capacidades de lenguaje sorprendentes
- **2022**: ChatGPT alcanza 100 millones de usuarios en 2 meses

```python
# Visualización del crecimiento exponencial en capacidades de IA
import matplotlib.pyplot as plt
import numpy as np

# Datos aproximados de capacidad computacional en entrenamiento
años = [2012, 2014, 2016, 2018, 2019, 2020, 2022]
petaflops_dias = [0.001, 0.01, 0.3, 3, 10, 314, 3640]  # Escala logarítmica aproximada
modelos = ['AlexNet', 'VGG', 'ResNet', 'BERT', 'GPT-2', 'GPT-3', 'PaLM']

plt.figure(figsize=(12, 6))
plt.semilogy(años, petaflops_dias, 'bo-', linewidth=2, markersize=8)

# Añadir etiquetas para cada punto
for i, modelo in enumerate(modelos):
    plt.annotate(modelo, (años[i], petaflops_dias[i]), 
                textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Año')
plt.ylabel('Petaflop-días de entrenamiento (escala log)')
plt.title('Crecimiento Exponencial en Capacidad Computacional de Modelos de IA')
plt.grid(True, alpha=0.3)
plt.show()

print("El crecimiento exponencial refleja la Ley de Moore aplicada a la IA")
```
